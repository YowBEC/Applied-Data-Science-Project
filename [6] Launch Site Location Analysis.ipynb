```python
import math
import builtins

# Define the Haversine function
def haversine(lon1, lat1, lon2, lat2):
    """
    Calculate the great circle distance in kilometers between two points
    on the earth (specified in decimal degrees).
    """
    # Convert decimal degrees to radians using builtins.map to avoid conflict
    lon1, lat1, lon2, lat2 = builtins.map(math.radians, [lon1, lat1, lon2, lat2])
    
    # Haversine formula
    dlon = lon2 - lon1 
    dlat = lat2 - lat1 
    a = math.sin(dlat/2)**2 + math.cos(lat1) * math.cos(lat2) * math.sin(dlon/2)**2
    c = 2 * math.asin(math.sqrt(a)) 
    r = 6371 # Radius of earth in kilometers
    return c * r

# Define proximity points (e.g., city coordinates)
proximities = {
    "City_A": (29.7604, -95.3698),  # Example coordinates (latitude, longitude)
    "City_B": (28.5383, -81.3792),
    "Landmark_C": (34.0522, -118.2437)
}

# Example launch site coordinates (latitude, longitude)
launch_sites = {
    "Launch_Site_1": (28.573255, -80.646895),
    "Launch_Site_2": (28.562302, -80.577356)
}

# Calculate distances
distances = {}
for site, (lat1, lon1) in launch_sites.items():
    distances[site] = {}
    for name, (lat2, lon2) in proximities.items():
        distance = haversine(lon1, lat1, lon2, lat2)
        distances[site][name] = distance

# Display the distances
distances

```

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

```python
import folium
import math

# Define the Haversine function
def haversine(lon1, lat1, lon2, lat2):
    """
    Calculate the great circle distance in kilometers between two points
    on the earth (specified in decimal degrees).
    """
    # Convert decimal degrees to radians
    lon1, lat1, lon2, lat2 = map(math.radians, [lon1, lat1, lon2, lat2])
    # Haversine formula
    dlon = lon2 - lon1
    dlat = lat2 - lat1
    a = math.sin(dlat/2)**2 + math.cos(lat1) * math.cos(lat2) * math.sin(dlon/2)**2
    c = 2 * math.asin(math.sqrt(a))
    r = 6371  # Radius of earth in kilometers. Use 3956 for miles. Determines return value units.
    return c * r

# Define Launch site coordinates
launch_site_lat = 28.573255
launch_site_lon = -80.646895

# Define the closest city coordinates
city_lat = 28.3922
city_lon = -80.6077

# Define the closest railway coordinates
railway_lat = 28.5721
railway_lon = -80.5853

# Define the closest highway coordinates
highway_lat = 28.57205
highway_lon = -80.58944

# Calculate the distance between the launch site and the closest city
distance_to_city = haversine(launch_site_lon, launch_site_lat, city_lon, city_lat)

# Calculate the distance between the launch site and the closest railway
distance_to_railway = haversine(launch_site_lon, launch_site_lat, railway_lon, railway_lat)

# Calculate the distance between the launch site and the closest highway
distance_to_highway = haversine(launch_site_lon, launch_site_lat, highway_lon, highway_lat)

# Create a folium map centered at the Launch site
site_map = folium.Map(location=[launch_site_lat, launch_site_lon], zoom_start=10)

# Create and add a folium.Marker for the closest city on the map
city_marker = folium.Marker(
    location=[city_lat, city_lon],
    icon=folium.DivIcon(
        icon_size=(20, 20),
        icon_anchor=(0, 0),
        html='<div style="font-size: 12; color:#d35400;"><b>%s</b></div>' % "{:10.2f} KM".format(distance_to_city),
    )
)
site_map.add_child(city_marker)

# Create and add a folium.Marker for the closest railway on the map
railway_marker = folium.Marker(
    location=[railway_lat, railway_lon],
    icon=folium.DivIcon(
        icon_size=(20, 20),
        icon_anchor=(0, 0),
        html='<div style="font-size: 12; color:#d35400;"><b>%s</b></div>' % "{:10.2f} KM".format(distance_to_railway),
    )
)
site_map.add_child(railway_marker)

# Create and add a folium.Marker for the closest highway on the map
highway_marker = folium.Marker(
    location=[highway_lat, highway_lon],
    icon=folium.DivIcon(
        icon_size=(20, 20),
        icon_anchor=(0, 0),
        html='<div style="font-size: 12; color:#d35400;"><b>%s</b></div>' % "{:10.2f} KM".format(distance_to_highway),
    )
)
site_map.add_child(highway_marker)

# Draw a PolyLine between the launch site and the closest city
city_lines = folium.PolyLine(locations=[[launch_site_lat, launch_site_lon], [city_lat, city_lon]], weight=1)
site_map.add_child(city_lines)

# Draw a PolyLine between the launch site and the closest railway
railway_lines = folium.PolyLine(locations=[[launch_site_lat, launch_site_lon], [railway_lat, railway_lon]], weight=1)
site_map.add_child(railway_lines)

# Draw a PolyLine between the launch site and the closest highway
highway_lines = folium.PolyLine(locations=[[launch_site_lat, launch_site_lon], [highway_lat, highway_lon]], weight=1)
site_map.add_child(highway_lines)

# Display the map
site_map
```
